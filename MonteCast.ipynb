```
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```
DATA_PATH = "https://raw.githubusercontent.com/FlaviaGoia/data_science/main/Monte_Castelo.csv"
```

```
df = pd.read_csv(DATA_PATH, encoding='latin-1')
```

```
df.head(25)
```

```
df = df.rename(columns={'Monte Castelo': 'montecastelo'})
```

```
df.head()
```

```
df.montecastelo.isnull().sum()
```

```
df
```

```
montecastelo = df['montecastelo']
```

```
all_montecastelo = " ".join(s for s in montecastelo)
print("Quantidade de Palavras: {}".format(len(all_montecastelo)))
```

```
df.head(50)
```

```
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black").generate(all_montecastelo)
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')

plt.tight_layout()
```

```
stopwords = set(STOPWORDS)
stopwords.update(["que", "em","de", "ou", "si", "ter", "um", "quer", "tão", "nos", "se"])
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1600, height=800).generate(all_montecastelo)
fig, ax = plt.subplots(figsize=(16,8))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
plt.imshow(wordcloud)
wordcloud.to_file("Monte_Castelo_montecastelo_wordcloud.png")
```
